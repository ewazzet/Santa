# Santa !

```python
import random

people = ['Madzia', 'Krzys', 'Ania', 'Janek', "Kasia", 'Ewa', 'Pawel & Sandra']

illegal_pairs = {
    'Madzia':'Krzys',
    'Krzys':'Madzia',
    'Ania':'Janek',
    'Janek':'Ania',

    'Madzia':'Kasia',
    'Janek':'Krzys',
    'Krzys':'Pawel & Sandra',
    'Pawel & Sandra':'Ewa',
    'Kasia':'Ania',
    'Ewa':'Madzia'
}

sneaky_santa = {}
```

```python
def draw_pair():
    santa = people[len(sneaky_santa)]
    child = people[random.randint(0, len(people)-1)]
    state = check(santa, child)
    if state:
      sneaky_santa.clear()
    elif santa not in sneaky_santa.keys():
        sneaky_santa[santa] = child
```

```python
def check(santa, child):
    if santa in illegal_pairs.keys():
        if child == illegal_pairs.get(santa):
            return True
    if santa == child:
        return True
    if child in sneaky_santa.values():
        return True
    return False
```

```python
def test(sneaky_santa):
    i = 0
    for person in people:
        if person not in sneaky_santa.keys() or person not in sneaky_santa.values():
          print('TEST FAILED !! {} not in the sneaky_santa'.format(person))
          i = i+1
        if people.count(person) > 1:
          print('TEST FAILED !! {} more then once in the sneaky_santa'.format(person))
          i=i+1
    if i == 0:
        print('TEST PASSED')
    else:
        print('THE WHOLE TEST WENT WROOOONG! TRY TO REPEAT! RUN AGAIN!')
```

```python
def main():
    while len(sneaky_santa) != len(people):
        draw_pair()
    test(sneaky_santa)
```

```python
if __name__ == '__main__':
    main()
```
